#### Trực quan hóa dữ liệu

Phân phối & Thống kê Dữ liệu

```python
# Lấy các thống kê mô tả cho tất cả các cột
all_column_stats = df_final.describe()

# Loại trừ giá trị 'count', 'min', 'max' khỏi thống kê cho tất cả các cột
all_column_stats = all_column_stats.drop(['count', 'min', 'max'])

# Tạo lưới 3x3 biểu đồ thanh (bar plots) bằng Plotly
fig = sp.make_subplots(rows=3, cols=3, subplot_titles=all_column_stats.columns)

# Lặp qua từng cột
for i, column in enumerate(all_column_stats.columns):
    row_num = i // 3 + 1
    col_num = i % 3 + 1

    # Chọn thống kê của cột
    column_stats = all_column_stats[column]

    # Thêm biểu đồ thanh vào biểu đồ con
    fig.add_trace(
        go.Bar(
            x=column_stats.index,
            y=column_stats.values,
            marker_color=px.colors.qualitative.Set3,
            texttemplate='%{y:.2s}',
            textposition='outside',
        ),
        row=row_num,
        col=col_num,
    )

    # Cập nhật layout của biểu đồ con
    fig.update_layout(
#         xaxis_title=f'Thống kê mô tả cho {column}',
        yaxis_title='Value',
    )

# Cập nhật layout cho toàn bộ biểu đồ
fig.update_layout(showlegend=False, height=500, width=900, title_text="Thống kê mô tả")
fig.show()

# Biểu đồ phân phối dữ liệu

columns = list(df_final.columns)[:9]

fig = sp.make_subplots(rows=3, cols=3, subplot_titles=[f'{column}' for column in columns])

for i, column in enumerate(columns):
    row, col = divmod(i, 3)  # Tính chỉ số hàng và cột
    histogram = go.Histogram(x=df_final[column], name=column)
    fig.add_trace(histogram, row=row + 1, col=col + 1)

fig.update_layout(height=500, width=950, title_text="Phân phối dữ liệu")
fig.update_xaxes(tickangle=45)  # Xoay nhãn trục x

fig.show()
```

Trực quan hóa phân phối dữ liệu và phát hiện các giá trị ngoại lai

```python
# Danh sách tên các cột số liệu
num_column_names = ['Volume', 'Open', 'Close', 'Adj Close', 'High', 'Low']

# Tạo lưới 2x3 để chứa các biểu đồ hộp (box plots)
fig = sp.make_subplots(rows=2, cols=3)

# Lặp qua từng cột số liệu để tạo biểu đồ hộp
for i, column in enumerate(num_column_names):
    # Tính toán chỉ số hàng và cột cho từng biểu đồ con
    row, col = divmod(i, 3)

    # Tạo biểu đồ hộp cho cột hiện tại
    box_trace = go.Box(y=df_final[column], name=column, jitter=0.3, pointpos=-1.8)

    # Thêm biểu đồ hộp vào vị trí tương ứng trong lưới
    fig.add_trace(box_trace, row=row + 1, col=col + 1)

# Cập nhật layout của biểu đồ tổng thể
fig.update_layout(height=500, width=950, title_text="Box Plots of Numeric Columns")

# Hiển thị biểu đồ
fig.show()
```

Phân tích Khối lượng Cổ phiếu Được Giao Dịch

Analysing the Prices of stocks of companies over the years

```python
def price(df_final, cmp_name, price_type):
    if price_type == 'Open':
        clr = 'orange'
    elif price_type == 'Close':
        clr = 'green'
    elif price_type == 'High':
        clr = 'blue'
    else:
        clr = 'purple'

    trace = go.Scatter(
        x=df_final.index,
        y=df_final[price_type],
        mode='lines',
        line=dict(color=clr),
        name=price_type,
        text=df_final['Date']  # Assuming 'Date' is the name of the date column in your DataFrame
    )

    layout = go.Layout(
        title=price_type + " Price of " + cmp_name,
        xaxis=dict(title='Date'),
        yaxis=dict(title=price_type)
    )

    fig = go.Figure(data=[trace], layout=layout)
    fig.show()
```

```python
price(df1,"Amazon","Open")
price(df2,"Apple","Open")
price(df3,"Facebook","Open")
price(df4,"Google","Open")
```

```python
price(df1,"Amazon","Close")
price(df2,"Apple","Close")
price(df3,"Facebook","Close")
price(df4,"Google","Close")
```

```python
price(df1,"Amazon","High")
price(df2,"Apple","High")
price(df3,"Facebook","High")
price(df4,"Google","High")
```

```python
price(df1,"Amazon","Low")
price(df2,"Apple","Low")
price(df3,"Facebook","Low")
price(df4,"Google","Low")
```

Phân tích khối lượng cổ phiếu được giao dịch

```python
def volume_analysis(df_final, cmp_name):
    trace = go.Scatter(x=df_final.index, y=df_final['Volume'],
                       mode='lines', marker=dict(color='#E493B3'))

    layout = go.Layout(title='Volume Of ' + cmp_name + ' Stock Prices',
                       xaxis=dict(title='Date'),
                       yaxis=dict(title='Volume'))

    fig = go.Figure(data=[trace], layout=layout)
    fig.show()
```

```python
volume_analysis(df1,'Amazon')
volume_analysis(df2,'Apple')
volume_analysis(df3,'Facebook')
volume_analysis(df4,'Google')
```

```python
df_final['Date'] = pd.to_datetime(df_final['Date'])
```

```python
df_Apple = df_final[df_final['Company'] == 'Apple']

# Assuming 'Date' column is in datetime format
# Extract year from the date
df_Apple['Year'] =df_Apple['Date'].dt.year

# Group by year and calculate the sum of volume for each year
df_Apple_yearly = df_Apple.groupby('Year')['Volume'].sum().reset_index()

# Plot the volume of stocks traded over time for Apple by year
fig = px.line(df_Apple_yearly, x='Year', y='Volume',
              labels={'Volume': 'Volume', 'Year': 'Year'},
              title='Volume of Stocks Traded Over Time for Apple by Year')

# Show the plot
fig.show()
```
